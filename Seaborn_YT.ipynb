**BOX PLOT**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,color="g",hue="sex",order=["Fri","Sun","Thur","Sat"],showmeans=True
            ,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette="plasma",orient="v")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="total_bill",y="day",data=var,color="g",hue="sex",order=["Fri","Sun","Thur","Sat"],showmeans=True
            ,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette="plasma")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=var,
            showmeans=True
            ,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette="plasma",orient="h")
plt.show()
```
