```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('petrol_consumption.csv')
df.head()
# BAR PLOT
df =df.dropna(subset =['Petrol_tax','Average_income','Paved_Highways','Population_Driver_licence(%)','Petrol_Consumption'])
Petrol_tax_counts= df['Petrol_tax'].value_counts()
plt.figure(figsize=(10,6))
plt.bar(Petrol_tax_counts.index,Petrol_tax_counts.values,color=['skyblue','orange'],width=0.2)
plt.title('Sales analysis')
plt.xlabel('Petrol_tax')
plt.ylabel('Count')
plt.xticks(np.arange(5.5,10,step=0.5),rotation=0)
plt.yticks(np.arange(2,20,step=2))
plt.tight_layout()
plt.savefig('bar_plot_petrol_consumption.png')
plt.show()

# HORIZONTAL BAR PLOT

df =df.dropna(subset =['Petrol_tax','Average_income','Paved_Highways','Population_Driver_licence(%)','Petrol_Consumption'])
Petrol_tax_counts= df['Petrol_tax'].value_counts()
plt.figure(figsize=(10,6))
plt.barh(Petrol_tax_counts.index.astype(str),Petrol_tax_counts.values,color=['skyblue','orange'],height=0.2) # Use barh for horizontal bar plot and height for bar thickness
plt.title('Sales analysis')
plt.xlabel('Count') # Swap xlabel and ylabel for horizontal bar plot
plt.ylabel('Petrol_tax') # Swap xlabel and ylabel for horizontal bar plot
# Adjusting ticks for horizontal plot. xticks now apply to the count axis.
plt.xticks(np.arange(0, Petrol_tax_counts.values.max() + 2, step=2))
# yticks now apply to the Petrol_tax categories.
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('horizontal_bar_plot_petrol_consumption.png') # Save with a new filename
plt.show()

# PIE PLOT

Average_income_counts = df['Average_income'].value_counts()
plt.figure(figsize=(10,6))
plt.pie(Average_income_counts.values, labels=Average_income_counts.index, autopct='%1.1f%%', shadow=True, radius=1.2, startangle=90)
plt.title('Average income Analysis')
plt.tight_layout()
plt.subplots_adjust(top=1.85)
plt.show()

# Histogram

df_cleaned = df.dropna(subset=['Petrol_Consumption'])

# Check if there's data to plot after cleaning
if not df_cleaned['Petrol_Consumption'].empty:
    plt.figure(figsize=(10, 6))
    plt.hist(df_cleaned['Petrol_Consumption'], bins=30, color='purple', edgecolor='black')

    # Corrected Title and Labels for a Petrol Consumption distribution histogram
    plt.title('Distribution of Petrol Consumption')
    plt.xlabel('Petrol Consumption')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()
else:
    print("No data to plot for 'Petrol_Consumption' after dropping NaN values.")

# Scatter Plot

plt.figure(figsize=(10, 6))
plt.scatter(df['Paved_Highways'], df['Petrol_Consumption']) # Assuming you want to scatter plot Paved_Highways against Petrol Consumption
plt.title('Paved Highways vs Petrol Consumption')
plt.xlabel('Paved Highways')
plt.ylabel('Petrol Consumption')
plt.show()
```
