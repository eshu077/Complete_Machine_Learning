```python
import numpy as np
np.random.seed(0)
data = np.random.normal(loc=0,scale=1,size=100)
len(data),data
```

```python
sns.boxplot(data, color='c')
plt.grid()
plt.show()
```

```python
sns.distplot(data, color='r')
plt.grid()
plt.show()
```

```python
# calculate IQR and Quartiles
```

```python
Q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
IQR = q3 - Q1
print(Q1,q3,IQR)
```

```python
# Define upper and lower bounds for outliers
upper_bound = Q1 - 1.5 * IQR
lower_bound = q3 + 1.5 * IQR

# Identify outliers

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Lower bound for Outliers:", lower_bound)
print("Upper bound for Outliers:", upper_bound)
print("Identify Outliers:", outliers)
```

```python
# Confidence Interval
```

```python
# Given parameters
n=40  # Sample size
x_bar=9.2  # Sample mean
s=1.8  # Sample standard deviation

# Generate Synthetic sample data using normal distribution
np.random.seed(42)  # For reproducibility

sample_data = np.random.normal(loc=x_bar, scale=s, size=n)
sample_data
```

```python
from scipy.stats import t

# step 1 : Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)   # Sample standard deviation
n = len(sample_data)
df = n-1
confidence_level = 0.95
alpha = 1 - confidence_level

# step 2 : Calculate the t-value
t_critical = t.ppf((1-alpha)/2, df)

# step 3 : Standard Error and Margin of error
SE  = sample_std / np.sqrt(n)
margin_of_error = t_critical * SE

# step 4 : Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

{
   "Sample Mean "  : round(sample_mean, 2),
   "Sample Standard Deviation" : round(sample_std, 2),
   "Sample Size" : n,
   "Degrees of Freedom" : df,
   "t-critical" : round(t_critical, 3),
   "Standard Error" : round(SE, 3),
   "Margin of Error" : round(margin_of_error, 3),

   "95% Confidence Interval" : (round(lower_bound, 3), round(upper_bound, 3))
}



```

```python
#step 5: plotting sample distribution and ci
CI_lower = lower_bound
CI_upper = upper_bound
plt.figure(figsize=(10,6))
plt.hist(sample_data,bins=10,color='skyblue',edgecolor='k',alpha=0.7)
plt.axvline(sample_mean,color='red',linestyle='dashed',label=f'Mean={sample_mean:.2f}')
plt.axvline(CI_lower,color='green',linestyle='dashed',label=f'95% CI_lower={CI_lower:.2f}')
plt.axvline(CI_upper,color='green',linestyle='dashed',label=f'95% CI_upper={CI_upper:.2f}')
plt.title('Histogram of Sample Data with Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Return ci values
(sample_mean, sample_std, (lower_bound, upper_bound))
```

```python
plt.figure(figsize=(10,6))
plt.plot(sample_data,marker='o',linestyle='-',color='darkblue',label='Battery life (hrs)')
plt.axhline(sample_mean, color='darkgreen', linestyle='dashed', label=f'Mean={sample_mean:.2f}')
plt.fill_between(range(n),CI_lower,CI_upper,color='g',alpha=0.2,label='95% CI')
plt.title("Battery life Sample data with 95% Confidence Interval")
plt.xlabel('Sample')
plt.ylabel('Battery life (hrs)')
plt.legend()
plt.grid(True)
plt.show()
plt.tight_layout()
```

Formulae for confidential interval (ci) using t-distribution ...**bold text**

```python
import numpy as np
from scipy.stats import t

# Sample data
n=40               # Sample size
x_bar=9.2          # sample mean battery life in hours
s=1.8              # sample standard deviation
confidential_level= 0.95
df = n-1           # degrees of freedom
df
```

```python
t_critical = t.pdf(1-(1-confidential_level)/2,df)
t_critical
```

```python
# Standard Error
SE = s / np.sqrt(n)

# Margin of error
ME = t_critical * SE

# Confidence Interval
CI_lower = x_bar - ME
CI_upper = x_bar + ME

{
    "Sample Mean(x)" : round(x_bar, 2),
    "Sample Standard Deviation(s)" : round(s, 2),
    "Sample Size(n)" : n,
    "Degrees of Freedom (df)" : df,
    "t-critical" : round(t_critical, 3),
    "Standard Error (SE)" : round(SE, 3),
    "Margin of Error (ME)" : round(ME, 3),
    "95% Confidence Interval" : (round(CI_lower, 3), round(CI_upper, 3))
}
```

```python
# Plot setup
plt.figure (figsize=(8,3))

# Plotting the confidence interval
plt.errorbar(x_bar,0,xerr=ME, fmt='o',color='blue',ecolor='red',elinewidth=2,capsize=5)

# Add annotations
plt.text(x_bar,0.05,f"Mean= {x_bar} hrs",ha = 'center',fontsize=10)
plt.text(CI_lower,-0.05,f"{CI_lower:2f}", ha='center', color='red')
plt.text(CI_upper,-0.05,f"{CI_upper:2f}", ha='center', color='green')

# Title and labels
plt.title('95% Confidence interval for batttery life')
plt.xlabel('Battery life (hrs)')
plt.yticks([]) # Hide y-axis
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
# Parameters
df=39
confidential_level=0.95
alpha = 1-confidential_level
t_critical= t.ppf(1-alpha/2,df)

# x values for the t-distibution curve

x= np.linspace(-4,4,100)
y=t.pdf(x,df)

print(x)
```

```python
# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x, y, label=f't-distribution (df={df})', color='blue')
plt.fill_between(x, y, where=(x < -t_critical), color='red', alpha=0.5, label='2.5% (left tail)')
plt.fill_between(x, y, where=(x > t_critical), color='red', alpha=0.5, label='2.5% (right tail)')
plt.axvline(-t_critical, color='black', linestyle='--', label=f'-t = {-t_critical:.3f}')
plt.axvline(t_critical, color='black', linestyle='--', label=f'+t = {t_critical:.3f}')
plt.title('95% Confidence Interval on t-distribution (df=39)')
plt.xlabel('t-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

**Hypothesis Test **


T- test

```python
from scipy import stats

wipro_salaries = [50000,55000,60000,65000,70000]
infosys_salaries = [45000,50000,55000,60000,65000]

alpha = 0.05   # Significance level

# Platform T-test
t_stat, p_value = stats.ttest_ind(wipro_salaries, infosys_salaries)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in salaries between Wipro and Infosys.")
else:
    print("Accept the null hypothesis. There is no significant difference in salaries between Wipro and Infosys.")
```
