```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Sample Dataset
```

```python
data = {

    'Independent_X': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'Label_Y': [3, 6, 19, 24, 45, 60, 85, 110, 150]

}

df = pd.DataFrame(data)
df
```

```python
plt.scatter(df['Independent_X'], df['Label_Y'], color='b')
plt.title('Independent_X vs Label_Y')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks(np.arange(1, 10, 1))
# plt.grid()
plt.show()
```

```python
# Separate independent and Dependent Data
```

```python
X = df[['Independent_X']]
y = df['Label_Y']
```

```python
X
```

```python
y
```

```python
# Model Training (Linear Regression)
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
model.coef_
```

```python
# y = mx + c
```

```python
# Model Prediction
```

```python
y_pred = model.predict(X)
```

**Ridge and Lasso Regression**

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('Melbourne_housing_cleaned.csv')
data
```

```python
X = data.drop('Price', axis=1)
y = data['Price']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test,y_test)
```

```python
pd.DataFrame(lr.coef_, index=X.columns, columns=['Coefficient/Slope/Weight'])
```

**Using Ridge Regression**

```python
from sklearn.linear_model import Ridge

```

```python
ridge = Ridge(alpha=1, max_iter=100, tol=0.1)
ridge.fit(X_train, y_train)
```

```python
ridge.score(X_train, y_train)
```

```python
ridge.score(X_test, y_test)
```

```python
# Len(ridge.coef_), ridge.coef_
```

```python
pd.DataFrame(ridge.coef_, index=X.columns, columns=['Coefficient/Slope/Weight'])
```

**Using Lasso Regression**

```python
# from sklearn import Linear_model
from sklearn.linear_model import Lasso
```

```python
lasso=Lasso(alpha=50, max_iter=100, tol=0.1)
lasso.fit(X_train, y_train)
```

```python
lasso.score(X_train, y_train)
```

```python
Lasso.score(X_test, y_test)
```

```python
# Lasso.coef_
```

```python
pd.DataFrame(lasso.coef_, index=X.columns, columns=['Coefficient/Slope/Weight'])
```
